```python
import pandas as pd

# Load the hourly ridership dataset
hour_df = pd.read_csv(r'C:\Users\venky\Downloads\archive\hour.csv')  # Replace with the actual path
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot Average Hourly Ridership (Total)
plt.figure(figsize=(10, 6))
sns.lineplot(x='hr', y='cnt', data=hour_df, estimator='mean', color='b')
plt.title('Average Hourly Ridership (Total)')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Total Riders')
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

```python
# Plot Casual vs Registered Riders by Hour
plt.figure(figsize=(10, 6))
sns.lineplot(x='hr', y='casual', data=hour_df, estimator='mean', label='Casual Riders', color='orange')
sns.lineplot(x='hr', y='registered', data=hour_df, estimator='mean', label='Registered Riders', color='green')
plt.title('Hourly Ridership: Casual vs Registered Riders')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Number of Riders')
plt.legend()
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```
